```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.read_csv(r'E:\Machine Learning\archive\Salary_Data.csv')
plt.title("Work experience vs salary prediction")
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()

X = df[['YearsExperience']].values
Y = df['Salary'].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.5, random_state=42)

model = LinearRegression()
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)
mse = mean_squared_error(Y_test, Y_pred)
print(f"MSE {mse}")

plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.plot(X_test, Y_pred,color = 'red' ,label = 'Predicted_values', marker = 'x')
plt.legend()
plt.show()  

plt.scatter(X_test, Y_test - Y_pred)
plt.hlines(y=0, xmin=X_test.min(), xmax=X_test.max(), colors='r')
plt.title('Residual Plot')
plt.xlabel('YearsExperience')
plt.ylabel('Residuals')
plt.show()

years_of_experience = np.array([[5]])
predicted_salary = model.predict(years_of_experience)
print(f"Predicted salary for 5 years of experience: {predicted_salary[0]}")
```
